# 1. Gravação de Áudio

```python
from IPython.display import Audio, display, Javascript
from google.colab import output
from base64 import b64decode

RECORD = """
const sleep = time => new Promise(resolve => setTimeout(resolve, time))
const b2text = blob => new Promise(resolve => {
  const reader = new FileReader()
  reader.onloadend = e => resolve(e.srcElement.result)
  reader.readAsDataURL(blob)
})

var record = time => new Promise(async resolve => {
  stream = await navigator.mediaDevices.getUserMedia({ audio: true })
  recorder = new MediaRecorder(stream)
  chunks = []
  recorder.ondataavailable = e => chunks.push(e.data)
  recorder.start()
  await sleep(time)
  recorder.onstop = async () => {
    blob = new Blob(chunks)
    text = await b2text(blob)
    resolve(text)
  }
  recorder.stop()
})
"""

def record(sec=15):
  display(Javascript(RECORD))
  js_result = output.eval_js('record(%s)' % (sec * 1000))
  audio = b64decode(js_result.split(',')[1])
  file_name = 'request_audio.wav'
  with open(file_name, 'wb') as f:
    f.write(audio)
  return f'/content/{file_name}'

print('Ouvindo...\n')
record_file = record()
display(Audio(record_file, autoplay=True))
```

### Opcional: Compactar o Notebook e o Arquivo de Áudio

```python
import os
import shutil

# Obtém o nome do notebook atual (se disponível)
# Em um ambiente Colab, o nome do arquivo .ipynb é geralmente acessível via 'sys.argv'
# ou pode ser inferido do ambiente.
# Para simplificar, vamos assumir que você fará o download manual do .ipynb
# e então executará este código para compactar.

# O arquivo de áudio já existe
audio_file = 'request_audio.wav'

# Nome do arquivo ZIP de saída
zip_filename = 'projeto_DIO.zip'

# Remove o arquivo zip existente se houver para evitar erros ou conteúdo antigo
if os.path.exists(zip_filename):
    os.remove(zip_filename)

# Cria um diretório temporário para os arquivos a serem zipados
temp_dir = 'temp_for_zip'
if not os.path.exists(temp_dir):
    os.makedirs(temp_dir)

# Copia o arquivo de áudio para o diretório temporário
if os.path.exists(audio_file):
    shutil.copy(audio_file, os.path.join(temp_dir, audio_file))
else:
    print(f"Aviso: Arquivo de áudio '{audio_file}' não encontrado para incluir no ZIP.")

# Instruções para incluir o notebook (assumindo que o usuário vai arrastar/colocar aqui ou baixar manualmente)
# Você precisaria subir seu arquivo .ipynb para a sessão antes de compactá-lo aqui
# Ou, como instruído acima, baixar o .ipynb manualmente e este código só lidaria com o .wav

# Para este exemplo, vamos zipar apenas o arquivo de áudio se o notebook for baixado separadamente.
# Se você quiser incluir o notebook, precisaria carregá-lo para este ambiente
# (por exemplo, arrastando para a barra lateral de arquivos do Colab).
# Ou, compactar manualmente após baixar ambos.

# Criar o arquivo zip a partir do diretório temporário
shutil.make_archive(zip_filename.replace('.zip', ''), 'zip', temp_dir)

# Baixa o arquivo ZIP
files.download(zip_filename)

# Limpa o diretório temporário
shutil.rmtree(temp_dir)

print(f'Arquivo {zip_filename} criado e baixado com sucesso! Certifique-se de incluir seu notebook .ipynb junto.')
```
